```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
db = pd.read_csv('database.csv')
df = db.sample(n=1000)
df.head()
```

```python
fig_dims = (9, 7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(x='Magnitude', data=df, ax=ax)
```

```python
fig_dims = (9, 7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.boxplot(x='Magnitude', data=df, ax=ax)
```

```python
df.plot.pie(y='Magnitude')
```
